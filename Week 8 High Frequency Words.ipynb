```python
import nltk
```

```python
#Alice's Adventures in Wonderland by Lewis Carroll
url = "carroll-alice.txt"
alice = nltk.corpus.gutenberg.words(url)

#remove punctuation and numerics
alice = [w for w in alice if w.isalpha()]

#strip out stop words or common words
from nltk.corpus import stopwords
alice = [w for w in alice if w not in stopwords.words('english')]
```

```python
#total words
total_words= len(alice)
total_words
```

```python
#total unique words
unique_words = set(alice)
len(unique_words)
```

```python
#Taking the most common words, how many unique words represent half of the total words in the corpus
```

```python
import pandas as pd
#words frequency table
freq_words = nltk.FreqDist(alice)

#top 200 words
top_200 = freq_words.most_common(200)
top_words = pd.DataFrame(top_200,columns=['word','count'])
top_words
```

```python
#calculate relative frequency
top_words['relative_freq'] = top_words['count']/(total_words)
top_words
```

```python
#Top 200 words graph
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20, 10))
freq_words.plot(200)
```
